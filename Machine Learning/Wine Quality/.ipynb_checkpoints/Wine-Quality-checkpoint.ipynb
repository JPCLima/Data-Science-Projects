# Red Wine Quality [Project 1]

The two datasets are related to red and white variants of the Portuguese "Vinho Verde" wine. For more details, consult the reference [Cortez et al., 2009]. Due to privacy and logistic issues, only physicochemical (inputs) and sensory (the output) variables are available (e.g. there is no data about grape types, wine brand, wine selling price, etc.).

These datasets can be viewed as classification or regression tasks. The classes are ordered and not balanced (e.g. there are much more normal wines than excellent or poor ones).

Input variables (based on physicochemical tests):

1 - fixed acidity

2 - volatile acidity

3 - citric acid

4 - residual sugar

5 - chlorides

6 - free sulfur dioxide

7 - total sulfur dioxide

8 - density

9 - pH

10 - sulphates

11 - alcohol

Output variable (based on sensory data):

12 - quality (score between 0 and 10)

## Model

* Define the X & y (target: quality)
* Split the data: 
    * 80% training set
    * 30% training set

```python
# Split data into X & y
X = df.drop("quality", axis = 1)
y = df["quality"]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y,
                                                    test_size=0.20, 
                                                    random_state = 42)
# Cheking the length
print(f'Train set: ({len(X_train)},{len(y_train)})\nTest set: ({len(X_test)},{len(y_test)})')
```

```python
# Create a function to fit and score models
def fit_and_score(models, X_train, X_test, y_train, y_test):
    
    # Make a dictionary to keep model scores
    model_scores = {}
    
    # Dimesion of the subplots
    f, axes = plt.subplots(1, 3, figsize=(20, 5), sharey='row')

    # Iterate though the models
    for i, (key, classifier) in enumerate(models.items()):
        
        # Calculate the y_pred
        y_pred = classifier.fit(X_train, y_train).predict(X_test)
        
        # Evaluate the model and append its score to model_scores
        model_scores[key] = classifier.score(X_test, y_test)
        
        # Calculate the confusion matrix
        cf_matrix = confusion_matrix(y_test, y_pred)
        disp = ConfusionMatrixDisplay(cf_matrix)
        disp.plot(ax=axes[i])
        disp.ax_.set_title(key)
        disp.im_.colorbar.remove()
        disp.ax_.set_xlabel('')

    f.text(0.4, 0.1, 'Predicted label', ha='left')
    plt.subplots_adjust(wspace=0.40, hspace=0.1)
    
    # Evaluate the model and append its score to model_scores       
    f.colorbar(disp.im_, ax=axes)
    plt.show()
    
    # Create a df to better visualization
    model_scores_df = pd.DataFrame(model_scores, index=["accuracy"])
        
    return model_scores_df
```

```python
def compare_models(model_scores_df):
    fig = plt.figure(figsize = (10,5))
    ax = sns.barplot(data=model_scores_df)   
    ax.set(ylabel='Accuracy (%)')    
    plt.show()     
```

### Model - Version 1

Run the models without parameters:

```python
# Models V1
models_V1 = {"Naive Bayes": GaussianNB(),
          "KNN": KNeighborsClassifier(),
          "Decision Tree Classifier": DecisionTreeClassifier()}

# Run the function to get all scores and confusion matrix
model_scores = fit_and_score(models=models_V1,
                             X_train=X_train,
                             X_test=X_test,
                             y_train=y_train,
                             y_test=y_test)

model_scores
```

```python
compare_models(model_scores)
```

We can see that the feature scaling improve the accuracy of the KNN model

### Model - Version 3

Improving the model - manually

**KNN** 

Find the best K

```python
# Lets's tune KNN
train_scores = []
test_scores = []

# Create a list of differnt values for n_neighbors
neighbors = range(1, 30)

# Setup KNN instance
knn = KNeighborsClassifier()

# Loop through different n_neighbors
for i in neighbors:
    knn.set_params(n_neighbors=i)
    
    # Fit the algorithm 
    knn.fit(X_train, y_train)
    
    # Update the training scores list
    train_scores.append(knn.score(X_train, y_train))
    
    # Update the test scores list
    test_scores.append(knn.score(X_test, y_test))
    
# Visualization of the best k
# Create another figure
plt.figure(figsize=(10, 6))
plt.plot(neighbors, train_scores, label="Train score")
plt.plot(neighbors, test_scores, label="Test score")
plt.xticks(np.arange(1, 30, 1))
plt.xlabel("Number of neighbors")
plt.ylabel("Model score")
plt.legend()

# Get the best param
params_knn = {neighbors[i]: test_scores[i] for i in range(len(neighbors))} 
best_param_knn = max(params_knn, key=params_knn.get) 


print(f"Maximum KNN score on the test data: {max(test_scores)*100:.2f}%\nBest K: {best_param_knn}")
```

With the graph above we can conclude that the best k is 1, with **62% of accuracy** better then the previous k=20 (knn by default) 

**Decision Tree Classifier**

```python
# Lets's tune Decision Tree Classifier
train_scores = []
test_scores = []

# Create a list of max_depths
max_depths = range(1, 15)

# Max score
highest_max_depths = 0

# Setup DecisionTreeClassifier instance
clf = DecisionTreeClassifier()

for max_depth in max_depths:
    clf.set_params(max_depth=max_depth)

    # Fit the algorithm 
    clf.fit(X_train, y_train)
    
    # Update the training scores list
    train_scores.append(clf.score(X_train, y_train))
    
    # Update the test scores list
    test_scores.append(clf.score(X_test, y_test))
    
# Visualization of the best max_depth
# Create another figure
plt.figure(figsize=(10, 6))
plt.plot(max_depths, train_scores, label="Train score")
plt.plot(max_depths, test_scores, label="Test score")
plt.xlabel("Max Depth")
plt.ylabel("Model score")
plt.legend()

# Get the best param
params_tree = {max_depths[i]: test_scores[i] for i in range(len(max_depths))} 
best_param_tree = max(params_tree, key=params_tree.get) 

print(f"Maximum max_depths score on the test data: {max(test_scores)*100:.2f}%\nBest max_depth: {best_param_tree}")
```

**Compare models**

```python
# Models V3
models_V3 = {"Naive Bayes": GaussianNB(),
          "KNN": KNeighborsClassifier(n_neighbors=best_param_knn),
          "Decision Tree Classifier": DecisionTreeClassifier(max_depth=best_param_tree)}

# Run the function to get all scores and confusion matrix
model_scores = fit_and_score(models=models_V3,
                             X_train=X_train,
                             X_test=X_test,
                             y_train=y_train,
                             y_test=y_test)

model_scores
```

```python
compare_models(model_scores)
```

### Model - Version 4

Improving the model using Grid Search CV

**KNN**

```python
grid_params = {
    'n_neighbors': range(1,100),
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']  
}

gs = GridSearchCV(
    KNeighborsClassifier(), 
    grid_params, 
    verbose=1, 
    cv=20, 
    n_jobs=-1)

# Fit the model
gs_results = gs.fit(X_train, y_train)

# Get the best estimor
knn_grid = gs_results.best_estimator_
```

**Compare models**

```python
# Models V4
models_V4 = {"Naive Bayes": GaussianNB(),
          "KNN": clf_grid,
          "Decision Tree Classifier": knn_grid}

# Run the function to get all scores and confusion matrix
model_scores = fit_and_score(models=models_V4,
                             X_train=X_train,
                             X_test=X_test,
                             y_train=y_train,
                             y_test=y_test)

model_scores
```
